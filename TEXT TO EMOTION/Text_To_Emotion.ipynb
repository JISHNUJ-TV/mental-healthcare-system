```python
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
train=pd.read_table('C:/Users/jishn/OneDrive/Desktop/jishnu/Emotion Detection/Dtaset/train.txt', delimiter = ';', header=None,)
val=pd.read_table('C:/Users/jishn/OneDrive/Desktop/jishnu/Emotion Detection/Dtaset/val.txt', delimiter = ';', header=None,)
test=pd.read_table('C:/Users/jishn/OneDrive/Desktop/jishnu/Emotion Detection/Dtaset/test.txt', delimiter = ';', header=None,)

data = pd.concat([train, val, test])
data.columns = ["text", "label"] 
```

```python
data.isna().any(axis=1).sum()
```

```python
#preprocessing

ps = PorterStemmer()

def preprocess(line):
    review = re.sub('[^a-zA-Z]', ' ', line)#leave characters only from a to z
    review = review.lower() #lower the text
    review = review.split() #turn the string into list of words
    #apply stemming
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')] #delete stopwords like I,AND ,OR etc
    #turn list into sentence
    return " ".join(review)
```

```python
data['text']=data['text'].apply(lambda x: preprocess(x))
```

```python
data
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
data['N_label'] = label_encoder.fit_transform(data['label'])
```

```python
data['text']
```

```python
#creating the bag of words model by applying countvectorizer -convert textual data in to numerical data
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,ngram_range=(1,3))
data_cv = cv.fit_transform(data['text']).toarray()
```

```python
data_cv
```

```python
#x_train, x_test, y_train, y_test = train_test_split(data_cv, data['N_label'], test_size=0.25, random_state=42)
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(data_cv, data['N_label'], test_size=0.25, random_state=42)
```

```python
#first neural network with keras tutorial


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

#load the dataset
#split into inputs(x) and output (y) variable
#define the keras model
model = Sequential()
model.add(Dense(12, input_shape=(x_train.shape[1], ), activation= 'relu'))
model.add(Dense(8, activation= 'relu'))
model.add(Dense(6, activation= 'softmax'))
#compile the keras model
model.compile(loss='sparse_categorical_crossentropy', optimizer= 'adam', metrics=['accuracy'])
#fit the keras model on the dataset
model.fit(x_train, y_train, epochs=10, batch_size=10)
#evaluate the kerasmodel


_, accuracy = model.evaluate(x_train, y_train)
print('Accuracy : %.2f' % (accuracy*100))
```

```python
_, accuracy = model.evaluate(x_test, y_test)
print('Accuracy : %.2f' % (accuracy*100))
```
